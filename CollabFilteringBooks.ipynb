```python
#Importing the library
import pandas as pd
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
#Loading the Data Set
books = pd.read_csv('Books.csv',low_memory=False)
ratings = pd.read_csv('Ratings.csv',low_memory=False)
users = pd.read_csv('Users.csv',low_memory=False)

ratings.head()
```

```python
books.head()
```

```python
ratings = pd.merge(books,ratings).drop(['Image-URL-L','Publisher','Image-URL-S','Image-URL-M','Year-Of-Publication'],axis=1)
ratings = pd.merge(users,ratings).drop(['Location','Age'],axis=1)
ratings.head(10)
```

```python
users.head()
```

```python
#Size of dataset
ratings.shape


users.shape
```

```python
#Author of The Da Vinci Code written in two different ways
ratings["Book-Author"]=ratings["Book-Author"].astype("string")
ratings["Book-Author"]=ratings["Book-Author"].str.replace("DAN BROWN","Dan Brown")
ratings["Book-Author"]=ratings["Book-Author"].astype("string")
ratings["Book-Author"]=ratings["Book-Author"].str.replace("JOHN GRISHAM","John Grisham")
```

```python
#Author of Dreamcatcher book correcting incorrect entries
ratings["Book-Author"]=ratings["Book-Author"].str.replace("Audrey Osofsky","Stephen King")
ratings["Book-Author"]=ratings["Book-Author"].str.replace("Dinah McCall","Stephen King")
```

```python
ratings.head(10)
```

```python
#Removing books with zero ratings from the data set
ratings=ratings[ratings["Book-Rating"]>0]
ratings["Book-Rating"].describe()
```

```python
ratings.head(10)
```

```python
#Unique reader count
ratings["User-ID"].nunique()
```

```python
#Unique number of books
ratings["Book-Title"].nunique()
```

```python
#How many times have we read which book?
book_counts = pd.DataFrame(ratings["Book-Title"].value_counts())

#Most read books
book_counts.sort_values("Book-Title", ascending=False)

#We named the books with less than 100 reads as rare books.
rare_book = book_counts[book_counts["Book-Title"] <= 50].index

#Number of rarely read books
rare_book.nunique()
```

```python
#By removing the rare books from the dataset, we found the widely read books
common_book = ratings[~ratings["Book-Title"].isin(rare_book)]
common_book.head()
```

```python
user_ratings = common_book.pivot_table(index=["User-ID"],columns=["Book-Title"],values="Book-Rating")
user_ratings.head(10)
```

```python
user_ratings.head(5)
```

```python
a= user_ratings.sort_values("User-ID", ascending=False)
a.head(5)
```

```python
#fill NA with 0
user_ratings= user_ratings.fillna(0)
```

```python
user_ratings.head(10)
```

```python
#Let's Build our similarity matrix
item_similarity_df = user_ratings.corr(method='pearson')
item_similarity_df.head(60)
```

```python
def get_similar_books(book_name,user_rating):
    similar_score = item_similarity_df[book_name]*(user_rating-5)
    similar_score = similar_score.sort_values(ascending=False)
    
    return similar_score
```

```python
sf = [("A Walk to Remember",10),("The Notebook",10),("Bridget Jones's Diary",9),
                 ("The Return of the King (The Lord of the Rings, Part 3)",2),("Lord of the Flies",2)]

similar_books = pd.DataFrame()
for book,rating in sf:
    similar_books = similar_books.append(get_similar_books(book,rating),ignore_index=True)

similar_books.head(50)
similar_books.sum().sort_values(ascending=False).head(20)
```
